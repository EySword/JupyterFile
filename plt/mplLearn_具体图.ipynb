```python
# %matplotlib qt5
# %matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
# matplotlib.pyplot.ion() #交互式开启
```

```python
#散点图

n=1024
X=np.random.normal(0,1,n)
Y=np.random.normal(0,1,n)
T=np.arctan2(Y,X) #color value

plt.scatter(X,Y,s=75,c=T,alpha=0.5)

plt.xlim((-1.5,1.5))
plt.ylim((-1.5,1.5))

plt.xticks(())
plt.yticks(())

plt.show()
```

```python
#柱状图

n=12
X=np.arange(n)
Y1=(1-X/float(n))*np.random.uniform(0.5,1.0,n)
Y2=(1-X/float(n))*np.random.uniform(0.5,1.0,n)

plt.bar(X,Y1,facecolor='#9999ff',edgecolor='white')
plt.bar(X,-Y2,facecolor='#ff9999',edgecolor='white')

for x,y in zip(X,Y1):
    plt.text(x,y+0.05,'%.2f'%y,ha='center',va='bottom') #ha横向对齐方式；va纵向对齐方式
    
for x,y in zip(X,-Y2):
    plt.text(x,y-0.05,'-%.2f'%y,ha='center',va='top') #ha横向对齐方式；va纵向对齐方式

plt.xlim(-.5,n)
plt.xticks(())
plt.ylim(-1.25,1.25)
plt.yticks(())

plt.show()
```

```python
#等高线图

def f(x,y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2) #算高度

n=256
x=np.linspace(-3,3,n)
y=np.linspace(-3,3,n)

X,Y=np.meshgrid(x,y) #定义表格 X:行点的每个x坐标 Y:每行点的每个Y坐标
print(type(X))

plt.contourf(X,Y,f(X,Y),8,alpha=0.75,cmap=plt.cm.hot) #cmap:地图颜色 8:分成8个高度
C=plt.contour(X,Y,f(X,Y),8,colors='black',linewidths=1) #边框
plt.clabel(C,inline=True,fontsize=10) #标注等高线高度

plt.xticks(())
plt.yticks(())
plt.show()
```

```python
#image 图片

a=np.array(np.random.random(9)).reshape(3,3)
print(a)

plt.imshow(a,interpolation='nearest',cmap='bone',origin='upper') #origin:数字大小对应颜色的规则  interpolation:输出样式
plt.colorbar(shrink=0.9) #图例，shrink为长度压缩值

plt.xticks(())
plt.yticks(())

plt.show()
```

```python
#3D 数据

from mpl_toolkits.mplot3d import Axes3D

fig=plt.figure()
ax=Axes3D(fig)
#定义XYZ
X=np.arange(-4,4,0.25)
Y=np.arange(-4,4,0.25)
X,Y=np.meshgrid(X,Y)
R=np.sqrt(X**2+Y**2)
Z=np.sin(R)

#3D图
ax.plot_surface(X,Y,Z,rstride=1,cstride=3,cmap=plt.get_cmap('rainbow')) #stride横轴间隔长度
#等高线图
ax.contourf(X,Y,Z,zdir='z',offset=-2,cmap='rainbow')  #zdir基于哪个轴投影  offset显示位置
ax.set_zlim(-2,2)

plt.show()
```
