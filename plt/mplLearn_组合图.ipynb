```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
# matplotlib.pyplot.ion() #交互式开启
```

```python
#subplot多合一显示

plt.figure()

plt.subplot(2,2,1) #创建小图 2行2列，第一个位置(下面代码对应的位置)
plt.plot([0,1],[0,1])

plt.subplot(2,2,2)
plt.plot([0,1],[0,2])

plt.subplot(2,2,3)
plt.plot([0,1],[0,3])

plt.subplot(2,2,4)
plt.plot([0,1],[0,4])

plt.show()
```

```python
#subplot 分格显示
#有三种方法

#1. subplot2grid
plt.figure()
ax1_1=plt.subplot2grid((3,3),(0,0),colspan=3,rowspan=1) #3行3列;从0行0列开始;跨度3列，一行
ax1_1.plot([1,2],[1,2]) #设置数据
ax1_1.set_title('ax1_title') #设置标题
ax1_2=plt.subplot2grid((3,3),(1,0),colspan=2)
ax1_3=plt.subplot2grid((3,3),(1,2),rowspan=2)
ax1_4=plt.subplot2grid((3,3),(2,0))
ax1_5=plt.subplot2grid((3,3),(2,1))
```

```python
#subplot 分格显示
#有三种方法

#2. gridspec
import matplotlib.gridspec as gridspec
plt.figure()
gs=gridspec.GridSpec(3,3) #设置总格数
ax2_1=plt.subplot(gs[0,:])
ax2_2=plt.subplot(gs[1,:2])
ax2_3=plt.subplot(gs[1:,2])
ax2_4=plt.subplot(gs[-1,0])
ax2_5=plt.subplot(gs[-1,-2])

plt.show()
```

```python
#subplot 分格显示
#有三种方法

#3. subplots
f,((ax11,ax12),(ax21,ax22))=plt.subplots(2,2,sharex=True,sharey=True)
ax11.scatter([1,2],[1,2])

plt.tight_layout()
plt.show()
```

```python
#plot in plot图中图

fig=plt.figure()
x=[1,2,3,4,5,6,7]
y=[1,3,4,2,5,8,6]

left,bottom,width,height=0.1,0.1,0.8,0.8  #图像显示的长宽，开始的位置
ax1=fig.add_axes([left,bottom,width,height])
ax1.plot(x,y,'r')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('title')

#直接创建第二个图
left,bottom,width,height=0.2,0.6,0.2,0.2
ax2=fig.add_axes([left,bottom,width,height])
ax2.plot(x,y,'b')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('title2')

#另一种创建的方式
plt.axes([.6,0.2,0.25,0.25])
plt.plot(y[::-1],x,'g')
plt.xlabel('x')
plt.ylabel('y')
plt.title('title2')

plt.show()
```

```python
#secondary axis 次坐标轴

x=np.arange(0,10,0.1)
y1=0.05*x**2
y2=-1*y1

fig,ax1=plt.subplots()
ax2=ax1.twinx() #y2坐标轴为y1的反向
ax1.plot(x,y1,'g-')
ax2.plot(x,y2,'b--')

ax1.set_xlabel('X data')
ax1.set_ylabel('Y1',color='g')
ax2.set_ylabel('Y2',color='b')

plt.show()
```

```python
#animation 动画
from matplotlib import animation

fig,ax=plt.subplots()

x=np.arange(0,2*np.pi,0.01)
line,=ax.plot(x,np.sin(x))

def animate(i):
    line.set_ydata(np.sin(x+i/100))
    return line,

def init():
    line.set_ydata(np.sin(x))
    return line,

#frames:100帧 init_func:最初的帧 interval:更新频率  blit:True只更新变化了的点
ani=animation.FuncAnimation(fig=fig,func=animate,frames=100,init_func=init,interval=20,blit=True)  

plt.show()
```
