```python
import h5py
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f = h5py.File("kappa5.hdf5",'r')
```

```python
g=f['gamma'][30]
g=np.where(g>0,g,-1)
lifetime=np.where(g>0,1.0/(2*2*np.pi*g),0)
```

```python
fre=f['frequency']
```

```python
print(lifetime.shape)
print(fre.shape)
```

```python
lifetime_df=pd.DataFrame(lifetime)
fre_df=pd.DataFrame(fre)
```
